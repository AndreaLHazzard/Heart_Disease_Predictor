```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
```

```python
#This dataset contains 303 rows and 14 columns
#sex(1 = male, 0 = female) 
#cp(chest pain type) 
#trestbps(resting blood pressure) 
#chol(cholesterol), fbs(fasting blood sugar) 
#restecg(resting electrocardiographic results) 
#thalach(maximum heart rate achieved) 
#exang(exercise induced angina) 1 = yes; 0 = no
#oldpeak(depression induced by exercise relative to rest) 
#slope(slope of the peak exercise ST segment) 
#ca(number of major vessels colored by fluoroscopy) 0-3 
#thal(thalassemia) 0= normal; 1 = fixed defect; 2 = reversible defect)
#condition(1 = heart disease, 0 = no heart disease)
heart = pd.read_csv('heart_cleveland_upload.csv')
heart = heart.dropna()
heart = heart.drop_duplicates()
heart.head()
```

```python
#train-test split, and visualize the data
x = heart.drop('condition', axis=1)
y = heart['condition']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#count occurrences of each class in each dataset
train_counts = y_train.value_counts().sort_index()
test_counts = y_test.value_counts().sort_index()

#set up bar positions
labels = ['No Disease (0)', 'Heart Disease (1)']
x = range(len(labels))
bar_width = 0.35

#plot bars
plt.figure(figsize=(8, 5))
plt.bar([i - bar_width/2 for i in x], train_counts, width=bar_width, label='Train', color='skyblue')
plt.bar([i + bar_width/2 for i in x], test_counts, width=bar_width, label='Test', color='salmon')

#labels and titles
plt.xticks(x, labels)
plt.xlabel('Heart Disease')
plt.ylabel('Count')
plt.title('Distribution of Target Variable in Train and Test Sets')
plt.legend()
plt.tight_layout()
plt.show()
```
